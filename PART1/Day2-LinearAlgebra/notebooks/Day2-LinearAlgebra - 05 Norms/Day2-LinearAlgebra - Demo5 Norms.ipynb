# 2.5 Norms

Normlar, aşağıdaki özelliklerle karakterize edilen işlevlerdir:

1- Normlar negatif olmayan değerlerdir. Normları bir uzunluk olarak düşünürseniz, neden negatif olamayacağını kolayca anlarsınız.

2- Normlar sadece ve sadece vektörün sıfır vektör olması durumunda $0$'dır.

3- Normlar üçgen eşitsizliğine saygı duyar.

4- $||{{k}\cdot {u}}||=|{k}| \cdot ||{{u}}||$. Bir vektörün bir skaler ile çarpımı normu, vektörün normu ile çarpılan bu skalerin mutlak değerine eşittir.

Genellikle iki yatay çubukla yazılır: $||{{x}}||$

# Üçgen eşitsizliği

Bazı vektörlerin toplamının normu, bu vektörlerin normlarının toplamından küçük veya ona eşittir.

$$
||{{u}+{v}}|| \leq ||{{u}}||+||{{v}}||
$$

### Örnek 1.

$$
{u}=
\begin{bmatrix}
    1 & 6
\end{bmatrix}
$$

ve

$$
{v}=
\begin{bmatrix}
    4 & 2
\end{bmatrix}
$$

$$
||{{u}+{v}}|| = \sqrt{(1+4)^2+(6+2)^2} = \sqrt{89} \approx 9.43
$$

$$
||{{u}}||+||{{v}}|| = \sqrt{1^2+6^2}+\sqrt{4^2+2^2} = \sqrt{37}+\sqrt{20} \approx 10.55
$$

# P-normları: genel kurallar

Bir vektörün $p$-normunu almanın tarifi:

1. Her elemanın mutlak değerini hesaplayın
2. Bu mutlak değerlerin $p$ kuvvetini alın
3. Tüm bu güçlü mutlak değerleri toplayın
4. Bu sonucun $\frac{1}{p}$ gücünü alın

Bu, şu formülle daha yoğun bir şekilde ifade edilir:

$$
||{{x}}||_p=(\sum_i|{x}_i|^p)^{1/p}
$$

Bu, yaygın olarak kullanılan bu $p$-normlarını kullanan örneklerle açıklığa kavuşacaktır.

# $L^1$ norm

$p=1$ yani bu norm basitçe mutlak değerlerin toplamıdır:

$$
||{{x}}||_1=\sum_{i} |{x}_i|
$$

# Euclidean norm ($L^2$ norm)

Öklid normu, $p=2$ olan $p$-normudur. Bu, kare $L^2$ normu ile daha çok kullanılan norm olabilir.

$$
||{{x}}||_2=(\sum_i {x}_i^2)^{1/2}\Leftrightarrow \sqrt{\sum_i {x}_i^2}
$$

Bu normun bir örneğini görelim:

### Örnek 2.

Grafik olarak, Öklid normu, vektörün orijinden doğrusal kombinasyonla elde edilen noktaya kadar olan uzunluğuna karşılık gelir (Pisagor teoreminin uygulanması gibi).

$$
{u}=
\begin{bmatrix}
    3 \\\\
    4
\end{bmatrix}
$$

$$
\begin{align*}
||{{u}}||_2 &=\sqrt{|3|^2+|4|^2}\\\\
&=\sqrt{25}\\\\
&=5
\end{align*}
$$


Yani $L^2$ normu $5$'dır.

# Squared Euclidean norm (squared $L^2$ norm)

$$
\sum_i|{x}_i|^2
$$


Squared $L^2$ normu uygundur çünkü karekökü kaldırır ve vektörün her kare değerinin basit toplamını elde ederiz.

Kare Öklid normu, kısmen ${x}^\text{T}{x}$ vektör işlemiyle hesaplanabildiğinden, makine öğreniminde yaygın olarak kullanılır. Optimizasyon nedeniyle performans artışı olabilir.

## Squared $L^2$ normunun türevi

Squared $L^2$ normunun bir başka avantajı, kısmi türevinin kolayca hesaplanmasıdır:

$$
u=
\begin{bmatrix}
    u_1\\\\
    u_2\\\\
    \cdots \\\\
    u_n
\end{bmatrix}
$$

$$
||{u}||_2 = u_1^2+u_2^2+\cdots+u_n^2
$$

$$
\begin{cases}
\dfrac{d||{u}||_2}{du_1} = 2u_1\\\\
\dfrac{d||{u}||_2}{du_2} = 2u_2\\\\
\cdots\\\\
\dfrac{d||{u}||_2}{du_n} = 2u_n
\end{cases}
$$

## $L^2$ normunun türevi

$L^2$ normu söz konusu olduğunda, türev daha karmaşıktır ve vektörün her öğesini hesaba katar:

$$
||{u}||_2 = \sqrt{(u_1^2+u_2^2+\cdots+u_n^2)} = (u_1^2+u_2^2+\cdots+u_n^2)^{\frac{1}{2}}
$$

$$
\begin{align*}
\dfrac{d||{u}||_2}{du_1} &=
\dfrac{1}{2}(u_1^2+u_2^2+\cdots+u_n^2)^{\frac{1}{2}-1}\cdot
\dfrac{d}{du_1}(u_1^2+u_2^2+\cdots+u_n^2)\\\\
&=\dfrac{1}{2}(u_1^2+u_2^2+\cdots+u_n^2)^{-\frac{1}{2}}\cdot
\dfrac{d}{du_1}(u_1^2+u_2^2+\cdots+u_n^2)\\\\
&=\dfrac{1}{2}\cdot\dfrac{1}{(u_1^2+u_2^2+\cdots+u_n^2)^{\frac{1}{2}}}\cdot
\dfrac{d}{du_1}(u_1^2+u_2^2+\cdots+u_n^2)\\\\
&=\dfrac{1}{2}\cdot\dfrac{1}{(u_1^2+u_2^2+\cdots+u_n^2)^{\frac{1}{2}}}\cdot
2\cdot u_1\\\\
&=\dfrac{u_1}{\sqrt{(u_1^2+u_2^2+\cdots+u_n^2)}}\\\\
\end{align*}
$$

$$
\begin{cases}
\dfrac{d||{u}||_2}{du_1} = \dfrac{u_1}{\sqrt{(u_1^2+u_2^2+\cdots+u_n^2)}}\\\\
\dfrac{d||{u}||_2}{du_2} = \dfrac{u_2}{\sqrt{(u_1^2+u_2^2+\cdots+u_n^2)}}\\\\
\cdots\\\\
\dfrac{d||{u}||_2}{du_n} = \dfrac{u_n}{\sqrt{(u_1^2+u_2^2+\cdots+u_n^2)}}\\\\
\end{cases}
$$

Squared $L^2$ normunun bir sorunu, fonksiyonun artışı yavaş olduğu için 0 ve küçük değerler arasında neredeyse hiç ayrım yapmamasıdır.

# Max norm

$L^\infty$ normudur ve vektörün en büyük öğesinin mutlak değerine karşılık gelir.

$$
||{{x}}||_\infty = \max\limits_i|x_i|
$$

# Matrix norms: Frobenius norm

$$

Bu, düzleştirmeden sonra matrisin $L^2$ normunu almaya eşdeğerdir.

# Nokta çarpımların normlarla ifadesi

$$
{x}^\text{T}{y} = ||{{x}}||_2\cdot||{{y}}||_2\cos\theta
$$


### Örnek 4.

$$
{x}=
\begin{bmatrix}
    0 \\\\
    2
\end{bmatrix}
$$

ve

$$
{y}=
\begin{bmatrix}
    2 \\\\
    2
\end{bmatrix}
$$

```python
x = [0,0,0,2]
y = [0,0,2,2]

plt.xlim(-2, 4)
plt.ylim(-2, 5)
plt.axvline(x=0, color='grey', zorder=0)
plt.axhline(y=0, color='grey', zorder=0)

plt.quiver([x[0], y[0]],
           [x[1], y[1]],
           [x[2], y[2]],
           [x[3], y[3]],
           angles='xy', scale_units='xy', scale=1)

plt.text(-0.5, 1, r'$\vec{x}$', size=18)
plt.text(1.5, 0.5, r'$\vec{y}$', size=18)

plt.show()
plt.close()
```

Bu örneği basitliği için aldık. Gördüğümüz gibi, $\theta$ açısı 45°'ye eşittir.

$$
{x^\text{T}y}=
\begin{bmatrix}
    0 & 2
\end{bmatrix} \cdot
\begin{bmatrix}
    2 \\\\
    2
\end{bmatrix} =
0\times2+2\times2 = 4
$$

ve

$$
||{{x}}||_2=\sqrt{0^2+2^2}=\sqrt{4}=2
$$

$$
||{{y}}_2=\sqrt{2^2+2^2}=\sqrt{8}
$$

$$
2\times\sqrt{8}\times cos(45)=4
$$

Numpy kullanan işlemler şunlardır:

```python
# Note: np.cos take the angle in radian
np.cos(np.deg2rad(45))*2*np.sqrt(8)
```
