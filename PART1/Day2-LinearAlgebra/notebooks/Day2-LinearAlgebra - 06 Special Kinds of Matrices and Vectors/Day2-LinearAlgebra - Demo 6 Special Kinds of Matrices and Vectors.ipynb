# Ortogonal vektörler

İki ortogonal vektör 90°'lik bir açıyla ayrılır. İki ortogonal vektörün nokta çarpımı 0 verir.

### Örnek 7.

```python
x = [0,0,2,2]
y = [0,0,2,-2]

plt.quiver([x[0], y[0]],
           [x[1], y[1]],
           [x[2], y[2]],
           [x[3], y[3]],
           angles='xy', scale_units='xy', scale=1)

plt.xlim(-2, 4)
plt.ylim(-3, 3)
plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')

plt.text(1, 1.5, r'$\vec{u}$', size=18)
plt.text(1.5, -1, r'$\vec{v}$', size=18)

plt.show()
plt.close()
```

## Özellik 1: ${A^\text{T}A}={I}$


Bir ortogonal matris şu özelliğe sahiptir:

$$
{A^\text{T}A}={AA^\text{T}}={I}
$$

Bu ifadenin doğru olduğunu şu mantıkla görebiliriz:

Aşağıdaki matrisi elde edelim:

$$
{A}=\begin{bmatrix}
    a & b\\\\
    c & d
\end{bmatrix}
$$

ve böylece

$$
{A}^\text{T}=\begin{bmatrix}
    a & c\\\\
    b & d
\end{bmatrix}
$$

Çarpmayı yapalım:

$$
\begin{align*}
&{A^\text{T}A}=\begin{bmatrix}
    a & c\\\\
    b & d
\end{bmatrix}
\begin{bmatrix}
    a & b\\\\
    c & d
\end{bmatrix}
=
\begin{bmatrix}
    aa + cc & ab + cd\\\\
    ab + cd & bb + dd
\end{bmatrix}\\\\
&=
\begin{bmatrix}
    a^2 + c^2 & ab + cd\\\\
    ab + cd & b^2 + d^2
\end{bmatrix}
\end{align*}
$$

[2.5](https://hadrienj.github.io/posts/Deep-Learning-Book-Series-2.5-Norms/)'de $\begin{bmatrix}
    a & c
\end{bmatrix}$ vektörünün normunun $a^2+c^2$ ($L^2$ or squared $L^2$)'a eşit olduğunu gördük. Ayrıca, ${A}$ dik olduğu için ${A}$ satırlarının bir birim normu olduğunu gördük. Bu, $a^2+c^2=1$ ve $b^2+d^2=1$ anlamına gelir. Şimdi elimizde:

$$
{A^\text{T}A}=
\begin{bmatrix}
    1 & ab + cd\\\\
    ab + cd & 1
\end{bmatrix}
$$

Ayrıca, $ab+cd$  $\begin{bmatrix}
    a & c
\end{bmatrix} ve \begin{bmatrix}
    b & d
\end{bmatrix}$'nin çarpımına eşittir:

$$
\begin{bmatrix}
    a & c
\end{bmatrix}
\begin{bmatrix}
    b\\\\
    d
\end{bmatrix}
=
ab+cd
$$

Ve sütunların ortogonal olduğunu biliyoruz, bu şu anlama gelir:

$$
\begin{bmatrix}
    a & c
\end{bmatrix}
\begin{bmatrix}
    b\\\\
    d
\end{bmatrix}=0
$$

Böylece birim matrisine sahibiz:

$$
{A^\text{T}A}=\begin{bmatrix}
    1 & 0\\\\
    0 & 1
\end{bmatrix}
$$

## Özellik 2: ${A}^\text{T}={A}^{-1}$

${A^\text{T}A}={I}$ ise $
{A}^\text{T}={A}^{-1}$ olduğunu gösterebiliriz.


${A^\text{T}A}={I}$ denkleminin her iki tarafını ${A}^{-1}$ ile çarparsak:

$$
({A^\text{T}A}){A}^{-1}={I}{A}^{-1}
$$

Bir matris veya vektör, birim matrisle çarpıldığında değişmez. Böylece:

$$
({A^\text{T}A}){A}^{-1}={A}^{-1}
$$

Matris çarpımı ilişkisel olduğundan parantezi kaldırabiliriz:

$$
{A^\text{T}A}{A}^{-1}={A}^{-1}
$$

Ayrıca biliyoruz ki, ${A}{A}^{-1}={I}$. Bu sebeple yer değiştirebiliriz :

$$
{A^\text{T}}{I}={A}^{-1}
$$

Bu şunu gösteriyor

$${A}^\text{T}={A}^{-1}$$
